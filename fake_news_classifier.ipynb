```python
import tensorflow as tf 
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM 
from tensorflow.keras.layers import Dense 
from tensorflow.keras.preprocessing.text import one_hot 
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

Preprocessing the text
re: Regressor is used for the String Manipulation 

nltk : Natural Language Toolkit(NLTK) is used for human language data

stopwords: the 'stopwords' corpus is collectoin of common words that are often considered to be littel value in text analysis , this are used to filter out the text because there are common words and don't have meaning, EG. "the", "and", "is", "in", etc

PortStemer: Provides a stemming algorithm that reduces words to their base or root form by removing suffixes. Reducing words to their base or root form to simplify text analysis

```python
import re 
import nltk 
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')
```

```python
import pandas as pd 
data = pd.read_csv('train.csv')
data.head(10)
```

Removing the NaN values from the dataset

It removing all the null values and reset the index value from the beginning 


```python
data = data.dropna()
data.reset_index(inplace=True)
```

Extracting the Dependant and Independant Variables from the data


```python
X = data.drop("label", axis=1)
y=data["label"]
```

```python
print(X.shape)
print(y.shape)
```

1 copying the train data to messages variable so no modifyling oocur for original
2 creating the object for PortStemmer
3 creating a empty list so that the preprocessed data will be added 
4 initalizing a for loop to iterate into entire message 
5 using review variable and re it will substitude the character with " " if incase not from a to z and this will be done in the messages['title']
6 lowering all characer 
7 spliting the words for stemming 
8 initializing another for loop if the words having the comming words like "but' "is" "and" will be removed using the for looop
9 joining all the words into the review 
10 appending with the empty list 


```python
messages = X.copy()
ps = PorterStemmer()
corpus = []
for i in range(len(messages)):
    review = re.sub("[^a-zA-Z]"," ", messages['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in  stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

One Hot Encoding 
processing each words by words with size of 5000 to process the onehot encoding 


```python
vocab_size =5000
onehot_repr = [one_hot(words,vocab_size) for words in corpus]
onehot_repr
```

Applying the Pad Seqences to make sentence length Equal
1  giving the maximum sentence length has 20 
2  using pad_sequences we are padding it with zero , (here padding='pre' indicate adding of 0 in front post means adding back with maxlength of 20 )

```python
sent_length=20
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

Creating a Model 


1 : This variable represents the dimensionality of the vector space in which words will be embedded 
2 : initializing a sequential model using keras. A sequenial Model is appropriate for a plain stack of layers where each layer has exactly one input tensor and one outpur tensor 
3 : Adds an embeding layer to model, (vocab_size) the number of uniqe words in the input data ,(embedded vector feature ) dimensity of the embedding vector , length of the input sequences, THis is crucial for LSTM 
4 : adding a LSTm with sigmoid which means output either 1 or 0 

```python
embedding_vector_features =40 
model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

Training and Testing the Data


```python
import numpy as np 
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X_final, y_final, test_size=0.3, random_state=42)
```

here we are fitting the training data with our model and then , giving a validation_data with the test data which is used to valid and give the accuracy of the data by performing it into our model , epochs has 10 , during the training data divided into smaller batches. The batch size is the number of training examples used in one iteration. 

```python
model.fit(X_train, y_train, validation_data=(X_test,y_test,), epochs=10, batch_size=64)
```

Performance Metrics and Accuracy


```python
y_pred = model.predict(X_test)
y_pred = np.round(y_pred).astype(int)
from sklearn.metrics import confusion_matrix 
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
